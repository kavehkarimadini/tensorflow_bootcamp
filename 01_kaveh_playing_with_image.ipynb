## <div style="direction:rtl;text-align:left;font-family:B Lotus, B Nazanin, Tahoma"> Gray scale image  </div>


<div style="direction:ltr;text-align:left;font-family:Tahoma">
the function below is for converting an rgb image to grayscaled image</div>
[stackoverflow](https://stackoverflow.com/questions/12201577/how-can-i-convert-an-rgb-image-into-grayscale-in-python)

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114]).astype(int)
```

```python
img = plt.imread('./image/t-shirt.jpg')
tshirt = rgb2gray(img)
plt.imshow(img)
```
